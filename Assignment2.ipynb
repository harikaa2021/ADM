```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Step 1: Load the dataset
df = pd.read_csv('employee_performance.csv')

# Display the dataset
print("Dataset:")
print(df.head())

# Step 2: Generate statistical summaries
print("\nStatistical Summaries:")
print(df.describe())

# Step 3: Data Visualization
# Histogram of Monthly Salaries
plt.figure(figsize=(10, 6))
sns.histplot(df['Monthly_Salary'], bins=20, kde=True)
plt.title('Histogram of Monthly Salaries')
plt.xlabel('Monthly Salary')
plt.ylabel('Frequency')
plt.show()

# Scatter Plot: Years of Experience vs. Performance Score
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Years_of_Experience', y='Performance_Score', data=df)
plt.title('Scatter Plot: Years of Experience vs. Performance Score')
plt.xlabel('Years of Experience')
plt.ylabel('Performance Score')
plt.show()

# Step 4: Data Transformation and Aggregation
# Example: Create a new derived column 'Salary_Experience_Ratio'
df['Salary_Experience_Ratio'] = df['Monthly_Salary'] / df['Years_of_Experience']

# Group by Department and calculate average salary and performance score
department_aggregation = df.groupby('Department').agg({
    'Monthly_Salary': 'mean',
    'Performance_Score': 'mean'
}).reset_index()

print("\nDepartment-wise Aggregation:")
print(department_aggregation)

# Save the transformed dataset to a new CSV file
df.to_csv('transformed_employee_performance.csv', index=False)
```
